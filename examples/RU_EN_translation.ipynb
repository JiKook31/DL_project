```
import torch

# List available models
torch.hub.list('pytorch/fairseq')
```

```
lines = []
with open("test.ru.txt", 'r') as f:
  raw_lines = f.readlines()
```

```
ru2en = torch.hub.load('pytorch/fairseq', 'transformer.wmt19.ru-en', checkpoint_file='model1.pt:model2.pt:model3.pt:model4.pt',
                       tokenizer='moses', bpe='fastbpe')
```

```
# Do not run this (not needed)
ru2en.eval()  # disable dropout
```

```
# Translate a sentence
ru2en.translate('С учетом изложенного выше Япония хотела бы предложить следующие элементы заключительных документов обзорной конференции 2015 года для дальнейшего обсуждения государствами-участниками.')
```

```
en_lines = []
i=0

for ru_line in raw_lines:
  en_line = ru2en.translate(ru_line)
  print(i)
  i+=1
  print(en_line)
  en_lines.append(en_line)
```
